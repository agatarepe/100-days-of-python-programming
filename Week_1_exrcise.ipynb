```python
class Robot:
    MAX_ENERGY = 100

    def __init__(self, name="Robot", age=25, energy = 0):  #

        self.name = name
        self.age = age
        self.energy = energy

    def display(self):
        print(f"I am {self.name}, age is {self.age}, energy is {self.energy}")


if __name__ == "__main__":
    robot = Robot()
    robot.display()





class Human:
    MAX_ENERGY = 100

    def __init__(self, name="Human", age=0):  #

        self.name = name
        self.age = age
        self.energy = Human.MAX_ENERGY

    def display(self):
        print(f"I am {self.name}")


if __name__ == "__main__":
    human = Human()
    human.display()
```
